```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Sample data
data = np.array([[25, 70], [30, 45], [35, 34], [40, 21], [45, 67],
                 [50, 34], [55, 98], [60, 65], [65, 10], [70, 55]])

# Scale data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=3)
labels = dbscan.fit_predict(scaled_data)

print(labels)
```

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Create synthetic data with clusters
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Add noise manually
noise = np.random.uniform(low=-10, high=10, size=(20, 2))
X_with_noise = np.vstack([X, noise])

# Fit KMeans on noisy data
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')
kmeans.fit(X_with_noise)
y_kmeans = kmeans.predict(X_with_noise)

# Plotting clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_with_noise[:, 0], y=X_with_noise[:, 1],
                hue=y_kmeans, palette='viridis', s=70, edgecolor='k', legend='full')

# Plot cluster centers
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')

# Title and Labels
plt.title('K-Means Clustering with Noise & Centroids', fontsize=14)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
